```python
print("List of unique values in column 'order_id': ")
order_id_unique = df['order_id'].unique()
print(order_id_unique)
print("Size:")
print(order_id_unique.size)
```

```python
# create a new list with 50 random items from order_id_unique,

random_order_id_unique = random.sample(list(order_id_unique), 500)

col_names = ["order_id", "tot_A_rows", "tot_E_rows", "tot_D_rows"]

# create a new dataframe with the column names, where column order_id are items in random_order_id_unique
df_random_order_id_unique = pd.DataFrame(columns = col_names)
df_random_order_id_unique['order_id'] = random_order_id_unique
```

```python
def count_tot_msg_type(order_id, msg_type):
    """
    Given order_id, returns the total number of rows where msg_type == A, E, D
    """
    tot_rows = df[(df.msg_type == msg_type) & (df.order_id == order_id)].shape[0]
    return tot_rows
```

```python
df_random_order_id_unique['tot_A_rows'] = df_random_order_id_unique.apply(lambda x: count_tot_msg_type(x.order_id, "A"), axis=1)
df_random_order_id_unique['tot_E_rows'] = df_random_order_id_unique.apply(lambda x: count_tot_msg_type(x.order_id, "E"), axis=1)
df_random_order_id_unique['tot_D_rows'] = df_random_order_id_unique.apply(lambda x: count_tot_msg_type(x.order_id, "D"), axis=1)
```

```python
df_random_order_id_unique

""""
Few findings:
1. Each msg_type except A can be zero -> Orders are assigned id at the first A order
2. Each msg_type can be more than 1
3. E may exceed sum of A and D -> Some order are executed twice, thus these should be ignored
4. D may not exceed A -> Each order A is cancelled at most once
5. A may not exceed sum of D and E, is usually the sum of D and E but sometimes can be less than that
-> Each order A is either executed or cancelled, but sometimes not cancelled
"""
```

```python
def get_order_sequence(order_id):
    """
    Given order_id, returns A list of msg_type in the order of their occurence
    """
    order_sequence = df[df.order_id == order_id].msg_type.to_list()
    return order_sequence

df_random_order_id_unique['order_list'] = df_random_order_id_unique.apply(lambda x: get_order_sequence(x.order_id), axis=1)
```
